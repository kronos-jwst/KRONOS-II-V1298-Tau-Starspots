```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## -- Visit 1 contrasts
# fleck
l1 = np.load('./data/contrasts/v1298tau_c_fleckcontrasts.npz')
waves1 = l1['waves']
wavehws1 = l1['wavehws']
cfleck1 = l1['C']
cerrfleck1 = l1['Cerr']

# gauss
l2 = np.load('./data/contrasts/v1298tau_c_gaussiancontrasts.npz')
cs1 = l2['C1']
cerrs1 = l2['C1err']
cs2 = l2['C1']
cerrs2 = l2['C1err']
cs3 = l2['C3']
cerrs3 = l2['C3err']
cs4 = l2['C4']
cerrs4 = l2['C4err']
cs5 = l2['C5']
cerrs5 = l2['C5err']
cs6 = l2['C6']
cerrs6 = l2['C6err']
cmed1 = l2['Ci_median']
cerrmed1 = l2['Ci_medianerr']

## -- Visit 2 contrasts
# fleck
waves2, wavehws2, cfleck2, cerrfleck2 = np.loadtxt('./data/contrasts/V1298Tau_contrasts_SOSSbd_fleck8_R100_v1.txt', skiprows=1, unpack=True)

# gauss
l4 = np.load('./data/contrasts/V1298Taubd_gauss8_contrasts.npz')

cs7 = l4['C7']
cerrs7 = l4['Cerr7']

cs8 = l4['C8']
cerrs8 = l4['Cerr8']

cs9 = l4['C9']
cerrs9 = l4['Cerr9']

cs10 = l4['C10']
cerrs10 = l4['Cerr10']

cs12 = l4['C12']
cerrs12 = l4['Cerr12']

cs14 = l4['C14']
cerrs14 = l4['Cerr14']

cmed2 = np.median((cs7,cs8,cs9,cs12,cs14), axis=0)
cerrmed2 = 1.2533 * (np.std((cs7,cs8,cs9,cs12,cs14),axis=0)/np.sqrt(6))

waves2 = np.delete(waves2, [20])
cfleck2 = np.delete(cfleck2, [20])
cerrfleck2 = np.delete(cerrfleck2, [20])
cs7 = np.delete(cs7, [20])
cerrs7 = np.delete(cerrs7, [20])
cs8 = np.delete(cs8, [20])
cerrs8 = np.delete(cerrs8, [20])
cs9 = np.delete(cs9, [20])
cerrs9 = np.delete(cerrs9, [20])
cs10 = np.delete(cs10, [20])
cerrs10 = np.delete(cerrs10, [20])
cs12 = np.delete(cs12, [20])
cerrs12 = np.delete(cerrs12, [20])
cs14 = np.delete(cs14, [20])
cerrs14 = np.delete(cerrs14, [20])
cmed2 = np.delete(cmed2, [20])
cerrmed2 = np.delete(cerrmed2, [20])
```

```python
fig, ax = plt.subplots(figsize=(6,12), nrows=3, sharex=True)
plt.subplots_adjust(hspace=0.15)
# top: visit 1
ax[0].set_title('Visit 1: 2025 Sept. 06', fontsize=12)
ax[0].errorbar(waves1, cs1, cerrs1, ls='None', marker='^', c='blue', lw=1, alpha=0.10, label='Spot 1', rasterized=True)
ax[0].errorbar(waves1, cs2, cerrs2, ls='None', marker='^', c='darkviolet', lw=1, alpha=0.10, label='Spot 2', rasterized=True)
ax[0].errorbar(waves1, cs3, cerrs3, ls='None', marker='^', c='mediumvioletred', lw=1, alpha=0.10, label='Spot 3', rasterized=True)
ax[0].errorbar(waves1, cs4, cerrs4, ls='None', marker='^', c='darkcyan', lw=1, alpha=0.10, label='Spot 4', rasterized=True)
ax[0].errorbar(waves1, cs5, cerrs5, ls='None', marker='^', c='darkgreen', lw=1, alpha=0.10, label='Spot 5', rasterized=True)
ax[0].errorbar(waves1, cs6, cerrs6, ls='None', marker='^', c='crimson', lw=1, alpha=0.10, label='Spot 6', rasterized=True)
ax[0].errorbar(waves1, cmed1, cerrmed1, ls='None', marker='^', c='black', mfc='white', lw=1, zorder=10, ms=4, label='Median', rasterized=True)
ax[0].errorbar(waves1, cfleck1, cerrfleck1, ls='None', marker='o', c='black', mfc='dimgray', lw=1, zorder=12, ms=4, label='fleck', rasterized=True)
ax[0].legend(loc='upper right', ncol=3)
ax[0].set_ylim(0,1)
ax[0].set_ylabel('Contrast', fontsize=12)

# middle: visit 2
ax[1].set_title('Visit 2: 2025 Sept. 10', fontsize=12)
ax[1].errorbar(waves2, cs7, cerrs7, ls='None', marker='^', c='blue', lw=1, alpha=0.10, label='Spot 7',rasterized=True)
ax[1].errorbar(waves2, cs8, cerrs8, ls='None', marker='^', c='darkviolet', lw=1, alpha=0.10, label='Spot 8',rasterized=True)
ax[1].errorbar(waves2, cs9, cerrs9, ls='None', marker='^', c='crimson', lw=1, alpha=0.10, label='Spot 9',rasterized=True)
ax[1].errorbar(waves2, cs10, cerrs10, ls='None', marker='^', c='mediumvioletred', lw=1, alpha=0.10, label='Spot 10',rasterized=True)
ax[1].errorbar(waves2, cs12, cerrs12, ls='None', marker='^', c='darkcyan', lw=1, alpha=0.10, label='Spot 12',rasterized=True)
ax[1].errorbar(waves2, cs14, cerrs14, ls='None', marker='^', c='darkgreen', lw=1, alpha=0.10, label='Spot 14',rasterized=True)
ax[1].errorbar(waves2, cmed2, cerrmed2, ls='None', marker='^', c='black', mfc='white', lw=1, zorder=10, ms=4, label='Median',rasterized=True)
ax[1].errorbar(waves2, cfleck2, cerrfleck2, ls='None', marker='o', c='black', mfc='dimgray', lw=1, zorder=12, ms=4, label='fleck',rasterized=True)
ax[1].legend(loc='upper right', ncol=3)
ax[1].set_ylim(0,1)
ax[1].set_ylabel('Contrast', fontsize=12)

# bottom: 
ax[2].errorbar(waves1, cfleck1, cerrfleck1, ls='None', marker='o', c='black', mfc='lightgray', lw=1, zorder=12, ms=4, label='fleck, Visit 1')
ax[2].errorbar(waves2, cfleck2, cerrfleck2, ls='None', marker='o', c='black', mfc='dimgray', lw=1, zorder=12, ms=4, label='fleck, Visit 2')
ax[2].legend(loc='upper right', ncol=3)
ax[2].set_ylim(0,1)
ax[2].set_ylabel('Contrast', fontsize=12)
ax[2].set_xlim(0.8, 2.8)
ax[2].set_xlabel(r'Wavelength ($\mu$m)', fontsize=12)

# h lines:
for i in range(3):
    ax[i].axvline(1.005, ls='--', c='gray', alpha=0.5, ymin=0, ymax=1) # Paschen n=7
    ax[i].axvline(1.094, ls='--', c='gray', alpha=0.5, ymin=0, ymax=1) # Paschen n=6
    ax[i].axvline(1.282, ls='--', c='gray', alpha=0.5, ymin=0, ymax=1) # Paschen beta (n=5)
    ax[i].axvline(1.875, ls='--', c='gray', alpha=0.5, ymin=0, ymax=1) # Paschen alpha
    ax[i].axvline(2.625, ls='--', c='gray', alpha=0.5, ymin=0, ymax=1) # Brackett beta (n=6)
    ax[i].axvline(2.166, ls='--', c='gray', alpha=0.5, ymin=0, ymax=1) # Brackett n=7
    ax[i].axvline(1.944, ls='--', c='gray', alpha=0.5, ymin=0, ymax=1) # Brackett n=8

plt.savefig('./figures/Figure4_obs_contrasts.pdf', dpi=600, bbox_inches='tight')
plt.show()
```
