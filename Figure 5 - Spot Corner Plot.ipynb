```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import corner
```

```python
with h5py.File('./data/fig5_selectspotproperties_flatsamples.h5', 'r') as f:
    fs_show = np.asarray(f['flatsamples'])
```

```python
fancylabels = np.array([r'$\alpha$',r'$r_{14}$',r'$lat_{14}$',r'$lon_{14}$'])

fig = plt.figure(figsize=(5,5))

figc = corner.corner(
    fs_show,
    labels=fancylabels,
    quiet=True,
    fig=fig,
    label_kwargs={"fontsize": 12},
    title_kwargs={"fontsize": 10},
    tick_params={"labelsize": 5},  
    labelpad=0.25,  
)

plt.savefig('./figures/Figure5_spotcorrelations.pdf', dpi=600, bbox_inches='tight')
plt.show()
```
